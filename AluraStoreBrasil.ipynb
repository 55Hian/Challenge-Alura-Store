### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Faturamento Total de Cada Loja
Faturamento1 = loja["Preço"].sum(numeric_only=True)
Faturamento2 = loja2["Preço"].sum(numeric_only=True)
Faturamento3 = loja3["Preço"].sum(numeric_only=True)
Faturamento4 = loja4["Preço"].sum(numeric_only=True)
```

```python
Faturamento4
```

# 2. Vendas por Categoria

```python
categorias1 = loja.groupby("Categoria do Produto")["Categoria do Produto"]
categorias2 = loja2.groupby("Categoria do Produto")["Categoria do Produto"]
categorias3 = loja3.groupby("Categoria do Produto")["Categoria do Produto"]
categorias4 = loja4.groupby("Categoria do Produto")["Categoria do Produto"]
```

```python
categorias4.value_counts().sort_values()
```

# 3. Média de Avaliação das Lojas


```python
avaliação1 = loja.groupby("Produto")["Avaliação da compra"].mean()
avaliação2 = loja2.groupby("Produto")["Avaliação da compra"].mean()
avaliação3 = loja3.groupby("Produto")["Avaliação da compra"].mean()
avaliação4 = loja4.groupby("Produto")["Avaliação da compra"].mean()

```

```python

avaliação2.sort_values()
```

# 4. Produtos Mais e Menos Vendidos

```python
vendaProdutos1 = loja.groupby("Produto")["Produto"]
maisVendidos1 = vendaProdutos1.value_counts().sort_values().tail()
menosVendidos1 = vendaProdutos1.value_counts().sort_values().head()

vendaProdutos2 = loja2.groupby("Produto")["Produto"]
maisVendidos2 = vendaProdutos2.value_counts().sort_values().tail()
menosVendidos2 = vendaProdutos2.value_counts().sort_values().head()

vendaProdutos3 = loja3.groupby("Produto")["Produto"]
maisVendidos3 = vendaProdutos3.value_counts().sort_values().tail()
menosVendidos3 = vendaProdutos3.value_counts().sort_values().head()

vendaProdutos4 = loja4.groupby("Produto")["Produto"]
maisVendidos4 = vendaProdutos4.value_counts().sort_values().tail()
menosVendidos4 = vendaProdutos4.value_counts().sort_values().head()
```
